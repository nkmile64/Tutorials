#### Graphs for Categorical Data (Nominal, Ordinal)

<dl>
    <dt style='font-weight:bold;color:green;'>Bar Chart</dt>
    <dd>
        - Each category is a separate bar<br/>
        - Bar height = frequency<br/>
        - Bars are separated (not touching)
    </dd>
    <dt style='font-weight:bold;color:red;'>Pie Chart</dt>
    <dd>
        - Shows parts of a whole<br/>
        - Best when we want to show proportions/percentages
    </dd>
</dl>

```python
# Python example: Bar Chart & Pie Chart
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D", "E"]
sizes = [10, 15, 7, 12, 9]

fig, axs = plt.subplots(1, 2, figsize=(9, 5), layout="constrained")

axs[0].bar(labels, sizes, color="lightblue", edgecolor="black")
axs[0].set_title("Bar Chart Example")

axs[1].pie(sizes, labels=labels, startangle=90, autopct="%1.1f%%")
axs[1].set_title("Pie Chart Example")
plt.show()
```

#### Graphs for Quantitative Data (Interval & Ratio)

<dl>
    <dt style='font-weight:bold;color:lime;'>Histogram</dt>
    <dd>
        - Bins (intervals) of group data e.g. 10-19, 20-29,...<br/>
        - Bars are adjacent to represent continuous data on the x-axis
    </dd>
</dl>

```python
# Python Example: Histogram
import matplotlib.pyplot as plt

data = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130, 111, 115, 112, 80, 75, 65, 54, 44, 43]
bins = [0, 20, 40, 60, 80, 100, 120, 140]

fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")
ax.hist(data, bins=bins, color="orange", edgecolor="black")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
ax.set_title("Histogram Example")

plt.show()
```

#### Types of Histogram Shapes

<p>
    <span style="color:cyan;font-weight:700;">Bell-Shaped (Normal)</span></br>
    Most data is in the center and few values at the extremes e.g. height, weight  
</p>
<p>
    <span style="color:red;font-weight:700;">Skewed Right (Positive Skew)</span></br>
    Long tail on the right e.g. income-most people earn little, few earn a lot 
</p>
<p>
    <span style="color:yellow;font-weight:700;">Skewed Left (Negative Skew)</span></br>
    Long tail on the left e.g. retirement age
</p>
<p>
    <span style="color:lime;font-weight:700;">Bimodal (Two Peaks)</span></br>
    Two distinct groups in the data e.g. test scores of two different classes.
</p>
