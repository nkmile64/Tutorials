```python
# The scores of a Math test taken by 10 students.
scores = [44, 47, 49, 51, 52, 53, 55, 57, 90, 95]

# Create a dot plot of the scores.
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 1.5), layout="constrained")
ax.plot(scores, [1] * len(scores), "ro")
ax.set_yticks([])
ax.set_xlabel("Score")
ax.grid(axis="x", color="lightgray", alpha=0.5)
ax.set_title("Dot Plot of Test Scores")

plt.show()
```

Two students (scoring 90 and 95) seem to have performed far better than the rest - possible **outliers**.

### STEP 3: Measuring the Center

- **Mode**: None (all values are unique)
- **Median**: There are two middle values so the median is the average of them:
    $= \frac{52 + 53}{2} = 52.5$
- **Mean**: $\bar{x} = \frac{\sum{x}}{n} = \frac{593}{10} = 59.3$

Notice that the mean is **pulled upward** from the median. This is because the mean is affected by the outliers (90 and 95), so in this case the median is a better measure of the center than the mean.

### STEP 4: How Spread Out are the Scores?

```python
import scipy.stats as stats

# Calculate the quartiles.
Q1 = stats.quantile(scores, 0.25)
Q3 = stats.quantile(scores, 0.75)
print(f"Q1 = {Q1}, Q3 = {Q3}")
```

- **Range**: $95 - 44 = 51$
- **IQR**: $Q3 - Q1 = 56.5 - 49.5 = 7$
- **Standard Deviation**: $\sigma = \sqrt{\frac{\sum{(x_i - \bar{x})^2}}{n}} = 16.8$

### STEP 5: Visualizing Spread with a Box Plot

```python
# fmt: off
import pandas as pd
import matplotlib.pyplot as plt

scores_ser = pd.Series(scores)

fig, ax = plt.subplots(figsize=(7, 3), layout="constrained")
bplot = ax.boxplot(
    scores_ser,                 # the sequence of values (grades)
    orientation="horizontal",   # plot the box horizontally
    sym="d",                    # symbol for flier points (diamond)
    widths=(0.8,),              # widths of the boxes
    patch_artist=True,          # draw box as a patch so as to set properties
)
bplot["boxes"][0].set_facecolor("orange")
bplot["fliers"][0].set_markeredgecolor("gray")
bplot["fliers"][0].set_markerfacecolor("gray")
bplot["medians"][0].set_color("black")

ax.grid(axis="x", color="lightgray", alpha=0.5)
ax.set_title("Box Plot of Test Scores")
ax.set_xlabel("Score")
ax.set_yticks([])

plt.show()
```

### STEP 6: Z-Score for the top student

The top student scored 95. How far is that from the mean?

$z = \frac{95 - 59.3}{16.8} = 2.1$

The top student scored 2.1 standard deviations above the mean - strong performance!

```python
# Compute the stats using Python
import numpy as np

scores = [44, 47, 49, 51, 52, 53, 55, 57, 90, 95]

mean = np.mean(scores)
median = np.median(scores)
std_dev = np.std(scores)
z_score = (95 - mean) / std_dev

print("Sample Statistics")
print("=" * 40)
print(f"Mean:\t\t\t{mean}")
print(f"Median:\t\t\t{median}")
print(f"Std Dev:\t\t{std_dev:.2f}")
print(f"Z-Score for 95:\t\t{z_score:.2f}")
```
