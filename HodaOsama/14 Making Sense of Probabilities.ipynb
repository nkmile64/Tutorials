# Union of Events: The Addition Rule

**The Addition Rule**:<br/>
The probability of **A or B** is the sum of the probabilities *minus* the overlap i.e. the intersection (because we added it twice):

<div style="text-align:center;">

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

<img alt="venn-addition-rule" src="img/venn_union_clear.png" style="width:80vw;">

</div>

**Disjoint Events**:<br/>
If A and B are mutually exclusive, then they cannot occur together i.e. $P(A \cap B) = 0$, so:

<div style="text-align:center;">

$P(A \cup B) = P(A) + P(B)$

</div>

# Conditional Probability

The probability of A **given** B i.e. when we know that B has already happened:

<div style="text-align:center;">

$\Large P(A|B) = \frac{P(A \cap B)}{P(B)}$

<img alt="conditional-prob" src="img/conditional_probability_venn.png" style="width:75vw;">

</div>

<span style="text-decoration:underline;font-style:italic;font-weight:500">Example: - Using the Table</span><br/>

What's the probability that a person owns a cat **given** they own a dog?

$P(Cat \space | \space Dog) = \frac{P(Cat \space and \space Dog)}{P(Dog)} = \frac{0.20}{0.45} = 0.44$
