### Regression: Predicting Relations Between Variables with the Best Fit Line

Once we establish the strength of the relationship between two variables with correlation, we use regression to model that relationship and make predictions.

We use Linear Regression to:
1. fit the best line through the data
2. calculate slope and intercept
3. assess prediction quality with residuals and R<sup>2</sup>

<h3 style="color:steelblue;">Predicting Exam Scores from Study Hours</h3>

Collecting data on student's study hours and exam scores, we find a Pearson's r:

r = 0.93

This indicates a strong positive correlation; more study hours result in higher grades.

<h3 style="color:lime">Drawing the Regression Line & Understanding Residuals</h3>

- Calculate the **residual** of each data point (the vertical distance to the regression line)
- **Positive residuals** are above the regression line
- **Negative residuals** are below the regression line

<h3 style="color:#DD1020">Ordinary Least Squares (OLS): Minimizing Residuals</h3>

The most common technique for fitting the regression line is OLS i.e. the line that minimizes the sum of the squared residuals:

<div style="text-align:center;">

$\Large \sum(y_i - \hat{y_i})^2$

</div>

<h3 style="color:orange;">The Regression Equation</h3>

<div style="text-align:center;">

$\Large \hat{y} = a + bx$

</div>

> $\hat{y}$ : the predicted value of y<br/>
  $a$ : the intercept (value of y when x = 0)<br/>
  $b$ : slope (the change in y for one unit increase in x)

<h3 style="color:#78AA67">Calculating the Regression Coefficients</h3>

<div style="text-align:center;">

$\Large b = r \times \frac{s_y}{s_x}$

$\Large a = \bar{y} - b\bar{x}$

</div>

> $r$ : Pearson's Correlation Coefficient between x and y<br/>
  $s_y$ : Standard Deviation of y<br/>
  $s_x$ : Standard Deviation of x<br/>
  $\bar{y}$ : Mean of y<br/>
  $\bar{x}$ : Mean of x

<h3 style="color:aqua;">Coefficient of Determination R<sup>2</sup> - Assessing the Fit</h3>

Measures:
1. How well the regression line predicts y
2. The proportion of variance in y, explained by x

<div style="text-align:center;">

$\Large R^2 = r^2$

</div>
