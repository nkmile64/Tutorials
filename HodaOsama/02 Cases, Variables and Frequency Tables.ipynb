**Cases** are the rows in a data table - each row is a case.<br/>
**Variables** are the characteristics we're measuring - they appear as columns.

#### Levels of Measurement

<dl>
    <dt style='font-weight:bold;color:green;'>Nominal</dt>
    <dd>Categories with no order: favorite subject</dd>
    <dt style='font-weight:bold;color:red;'>Ordinal</dt>

#### Frequency Table
It shows how often something appears in our data.

#### Hands-On Practice

```python
# Practice: Frequency Tables in Python
# This notebook will help you practice creating frequency tables from raw data,
# just like we discussed in the blog post.

import pandas as pd

# Example: Heights of students
data = {"Height": [150, 160, 165, 170, 160, 175, 180, 165, 170, 155, 160, 165, 170, 160, 175, 180, 165, 170, 155, 160]}
df = pd.DataFrame(data)
df.head()
```

```python
# Frequency table for heights (grouped by value)
freq_table = df["Height"].value_counts().sort_index()
print(freq_table)
```

```python
# Group heights into bins
bins = [140, 150, 160, 170, 180, 190]
labels = ["140-149", "150-159", "160-169", "170-179", "180-189"]
df["HeightGroup"] = pd.cut(df["Height"], bins=bins, labels=labels, right=False)
grouped_freq = df["HeightGroup"].value_counts().sort_index()
print(grouped_freq)
```
