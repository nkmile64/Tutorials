#### STEP 3: Understanding Proportions - Quick Summary

- **Conditional Proportions** : look within groups<br/>
    e.g.among those who study rarely only 30% passed the exam 
- **Marginal Proportions** : summarize a variable on its own<br/>
    e.g.overall pass rate = 18/30 = 60%

#### STEP 4: Interpreting the Categorical Correlation

We can see that there is a **positive association** in the conditional proportions:<br/>
    The more a student studies, the more likely they are to pass.

#### STEP 6: Scatter Plot (for Quantitative variables)

```python
import matplotlib.pyplot as plt
import numpy as np

hours = np.array([2, 3, 5, 7, 8, 10, 12])
score = np.array([50, 55, 65, 70, 76, 85, 92])

fig, ax = plt.subplots(figsize=(9, 6))

ax.scatter(hours, score, marker="x", s=90)
ax.plot(hours, score, "b--")
ax.set_title("Scatter Plot: Study Hours vs. Quiz Score", fontsize=16)
ax.set_xlabel("Hours Studied", fontsize=15)
ax.set_ylabel("Quiz Score", fontsize=15)
ax.grid(linestyle="dashed", alpha=0.5)
ax.set_frame_on(False)

plt.show()
```

So, the more hours a student studies, the higher their scores.<br/>
This is a **strong positive relationship**.
