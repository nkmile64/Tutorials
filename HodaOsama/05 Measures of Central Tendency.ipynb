The three main measures of **central tendency** are:
- **Mode** : the most frequent value in a dataset
- **Median** : the middle value (or the avg of the two middle values), when data is sorted
- **Mean** : the arithmetic average: add up all the values, and divide by the count

$\large Mean = \bar{x} = \frac{1}{n}\sum\limits_{i=1}^{n} {x_i}$

```python
from scipy import stats

data = [80, 90, 85, 90, 95, 90, 92]
modeResult = stats.mode(data, keepdims=True)

# The returned value of calling mode(), is a ModeResult object with
# mode and count keys. Each key is an array (since there might be more)
# than one mode, so we access the values with indexing:
print(f"Mode: {modeResult.mode[0]} appearing {modeResult.count[0]} times")
```

```python
import numpy as np

data = [16, 17, 18, 40, 105]
median = np.median(data)

print(f"Median: {median}")
```

```python
import numpy as np

scores = [80, 85, 90]
mean = np.mean(scores)

print(f"Mean: {mean}")
```

Note that the mean may be affected by outliers, thus giving wrong impressions. In that case, use the median.
