### Why use Variance and Standard Deviation?

- They show how **spread** out the data is from the mean
- And they use **every single value** to calculate that spread

### Variance Formula
#### Population Variance

$\Large\sigma^2 = \frac{1}{N}\sum\limits_{i=1}^{N}(x_i - \mu)^2$

Where:

- $x_i$ =  each individual value
- $\mu$ =  the population mean
- N     =  the population size

#### Sample Variance

$\Large s^2 = \frac{1}{n - 1}\sum\limits_{i=1}^{n}(x_i - \bar{x})^2$

Where:

- $x_i$ =      each individual value
- $\bar{x}$ =  the sample mean
- n     =      the sample size

### Why not use Raw Deviations?

Because, that would give us a result of zero!

### Step-by-Step Variance Example

For these five values, calculate the variance:<br/>
[4, 5, 7, 10, 14]

First, compute the mean:

$\bar{x} = \frac{4+5+7+10+14}{5} = 8$

So the population variance is: $\frac{66}{5} = 13.2$

Calculate the population variance using Python:

```python
import numpy as np

data = [4, 5, 7, 10, 14]

# Population Variance
# For sample variance use ddof=1
variance = np.var(data)
sd = np.std(data)
print(f"Variance \t\t: {variance}")
print(f"Standard Deviation \t: {sd:.2f}")
```

### Standard Deviation

Since variance is in squared units, we take the square root of the variance to calculate the **standard deviation**

$\sigma = \sqrt{\sigma^2} = \sqrt{13.2} = 3.63$

This tells us that the **average distance from the mean** is 3.63 units.
