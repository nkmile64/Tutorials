### Measuring Randomness with Probability

Probability is a number between 0 and 1 that describes the **likelihood** of an event.

$0 < P(event) < 1$

- P = 0.0 impossible
- P = 1.0 certain
- P = 0.5 equal chance

### Relative Frequency & Cumulative Proportion

For events that have already happened and we have observed them e.g. tossing the coin 1000 times, we can calculate probability with **relative frequency**:

$P(event) = \frac{event \space count}{total \space trials}$

As the number of trials increases, the running total of relative frequencies is called **cumulative proportion**; this gives us a better estimate of the true probability.

### Event Types

> **Mutually Exclusive (Disjoint)**<br/>
    Events that cannot happen at the same time (do not overlap)<br/>
    $P(A \cap B) = 0$

> **Collectively Exhaustive**<br/>
    Events that together cover all possibilities<br/>
    $\sum{P(events)} = 1$

> **Complement Event**<br/>
    All outcomes not in the event<br/>
    $P(A^c) = 1 - P(A)$
