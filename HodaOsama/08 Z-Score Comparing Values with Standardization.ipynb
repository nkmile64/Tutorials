> ***Z-score:*** how far the value is from the mean in terms of standard deviations
>
> They are used for:
> - Standardize features
> - Detect outliers
> - Compare different variables
> - Improve features scaling & model convergence

### Z-score standardization into the ML pipeline

1. For each numeric feature estimate the mean and the standard deviation of only the `training data`
2. Standardize the training features 

    $\Large z = \frac{x-\bar{x}}{s}$

3. Use the same mean/standard deviation from the training data, to transform the `test data`.

This way we avoid `data leakage` i.e. information from the test data, leaking into the training process.

### Using z-scores for outlier detection

- |z| <= 2 a typical value
- 2 < |z| <= 3 possible outlier
- |z| > 3 strong candidate for outlier

### Z-scores, Probabilities & Percentiles

For a standardized variable that follows the **standard normal distribution** N(0, 1):
> **Cumulative Probability** P(Z <= z) is the area under the curve to the left of z.

<dl>
    <dt style='font-weight:bold;color:red;'>z = 0</dt>
    <dd>50% of the values are below the mean</dd>
    <dt style='font-weight:bold;color:red;'>z = 1</dt>
    <dd>84% of the values are below the value</dd>
    <dt style='font-weight:bold;color:red;'>z = 2</dt>
    <dd>97.7% of the values are below the value</dd>
</dl>

> A **percentile** gives you the percentage of observations that fall at or below a value.

<dl>
    <dt style='font-weight:bold;color:lime;'>z = 0</dt>
    <dd>The value is at the 50th percentile</dd>
    <dt style='font-weight:bold;color:lime;'>z = 1.64</dt>
    <dd>The value is at the 95th percentile</dd>
    <dt style='font-weight:bold;color:lime;'>z = -1.64</dt>
    <dd>The value is at the 5th percentile</dd>
</dl>

### ML Example: Standardizing a feature before modelling

Build a model to predict whether a patien has **high** cardiovascular risk.<br/>
One of the features is cholesterol with:<br/>

$\large \bar{x} = 200mg/dL$ <br/>
$\large s = 40mg/dL$

A patient has cholesterol x = 260 mg/dL.<br/>
The z-score is:

$\Large z = \frac{x - \bar{x}}{s} = \frac{260 - 200}{40} = 1.5$

The rule of 68%-95%-99.7%, only applies when the distribution is close to normal, and not when the data are strongly skewed.
