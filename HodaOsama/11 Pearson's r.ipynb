#### **Pearson's Correlation Coefficent (r)** 
It measures how strongly are two variables related and in what direction.<br/>
As such, **r** quantifies the correlation that a **scatter plot** may indicate it exists between the variables.

> **NOTE**: If the relationship between the variables is not linear, the r coefficient may not be applicable.

#### The Equation for Pearson's r
<div style="text-align:center;">

$\LARGE r = \frac{\sum(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i - \bar{x})^2 \sum(y_i - \bar{y})^2}}$

</div>
where:

- $x_i$ = data points of variable x
- $\bar{x}$ = the mean of variable x
- $y_i$ = data points of variable y
- $\bar{y}$ = the mean of variable y

#### Python Code for Pearson's r

```python
import numpy as np

# data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 6, 8, 10])

# calculate Pearson's r
# corrcoef returns a correlation matrix (2x2 ndarray) so we index it
r = np.corrcoef(x, y)[0, 1]

print(f"Pearson's r: {r:.2f}")
```

This means the correlation is **very strong positive**.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(x, y)
ax.scatter(x, y, c="green", marker="*", s=60)
```
