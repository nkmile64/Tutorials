### Descriptive Statistics : "What do I see?"

```python
import numpy as np
import matplotlib.pyplot as plt

scores = [75, 88, 92, 60, 79, 85, 90, 70]

print(f"Mean\t\t\t:\t{np.mean(scores)}")
print(f"Standard Deviation\t:\t{np.std(scores)}")
print(f"Minimum\t\t\t:\t{np.min(scores)}")
print(f"Maximum\t\t\t:\t{np.max(scores)}")

plt.hist(scores, bins=5, color="skyblue", edgecolor="black")
plt.title("Exam Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```
