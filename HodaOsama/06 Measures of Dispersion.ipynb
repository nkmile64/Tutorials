### Range
Range = Largest value - Smallest value

$Range = x_{max} - x_{min}$

NOTE: If there are outliers, range is not a reliable measure of variability

### Interquartile Range (IQR)

<dl>
    <dt style='font-weight:bold;color:green;'>Q1</dt>
    <dd>25% of data is below this point</dd>
</dl>
<dl>
    <dt style='font-weight:bold;color:green;'>Q2</dt>
    <dd>50% of data is below this point (median)</dd>
</dl>
<dl>
    <dt style='font-weight:bold;color:green;'>Q3</dt>
    <dd>75% of data is below this point</dd>
</dl>

$IQR = Q_3 - Q_1$

IQR removes the influence of extreme values, by focusing on the **middle 50%** of the data.<br/>
The middle 50% of the data spans IQR units.

### Box Plot
The box plot summarizes:
- The min and max values (excluding outliers)
- Q1, Q2 (median) and Q3
- Any outliers (points beyond 1.5 x IQR from quartiles)
