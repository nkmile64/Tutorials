# **Independent Events**

Two events are **independent**, when the occurence of one, **doesn't affect** the probability of the other:

$P(A|B) = P(A)$

$P(B|A) = P(B)$

$P(A \cap B) = P(A) \times P(B)$ [**Multiplication Rule**]

<dl>
    <dt style='font-weight:bold;color:red;'>Disjoint (Mutually Exclusive):</dt>
    <dd>
        Events cannot both happen. Disjoint events are always <b>dependent</b>.
    </dd>
    <dt style='font-weight:bold;color:green;'>Independent:</dt>
    <dd>One event doesn't affect the other's probability.</dd>
</dl>

<span style="color:darkslateblue;font-weight:bold;font-style:italic;text-decoration:underline double;">Example:</span><br/>
We want to construct an email spam detection scheme so that we can answer the question:<br/>
> <span style="color:#898;">If we see the word "free" in an email, what's the probability it's spam?</span>

The following data are available:

- 40% of all emails are spam.
- 80% of spam emails, contain the word "free"
- 10% of non-spam emails contain the word "free"

Using the above data, we first draw a **decision tree**.

<img src="img/bayes_decision_tree.png" alt="decision tree" style="width:80vw;">

Since "spam email" and "containing the free word" are **independent events**, we can use the **multiplication rule** to calculate:

- P(Spam and Free) = 0.4 X 0.8 = 0.32
- P(Not Spam and Free) = 0.6 X 0.1 = 0.06

Now, calculate all **joint probabilities**:
- $P(S \cap F) = 0.4 \times 0.8 = 0.32$ - contains "free" and it's spam
- $P(\overline{S} \cap F) = 0.6 \times 0.1 = 0.06$ - contains "free" and it's not spam
- $P(F) = 0.32 + 0.06 = 0.38$ - all cases of containing "free"

So, to answer the question in the beginning we turn to <a href="./14 Making Sense of Probabilities.ipynb">**conditional probabilities**</a> to calculate the probability that an email is spam, given that it contains the word "free":

$\Large P(S | F) = \frac{P(S \ \cap \ F)}{P(F)} = \frac{0.32}{0.38} = 0.842$

#### Baye's Rule: Components

<dl>
    <dt style='font-weight:bold;color:green;'>Prior</dt>
    <dd>
        What you believe before seeing the evidence.<br/>
        P(S) = 0.4
    </dd>
    <dt style='font-weight:bold;color:red;'>Likelihood</dt>
    <dd>
        Probability of the evidence, given the hypothesis</br>
        P(F | S)
    </dd>
    <dt style='font-weight:bold;color:dardcyan;'>Evidence</dt>
    <dd>
        Total probability of seeing the evidence</br>
        P(F) = 0.38
    </dd>
    <dt style='font-weight:bold;color:brown;'>Posterior</dt>
    <dd>
        Updated belief</br>
        P(S | F) = 0.842
    </dd>
</dl>

<span style="color:darkslateblue;font-weight:bold;font-style:italic;text-decoration:underline double;">Example:</span><br/>
We want to construct a student cheating detection scheme, using a plagiarism detector that flags students:<br/>
> <span style="color:#898;">If a student gets flagged, what are the chances they actually cheated?</span>

The following data are available:
- Only 2% of students cheat on exams.
- Cheaters are correctly identified 90% of the time
- The detector wrongly flags innocent students 5% of the time.

Since "cheater student" and "student flagged" are two independent events, we can use the **mutiplication rule**, to calculate **joint probabilities**:

- $P(C \cap P) = P(C) \ P(P) = 0.02 \times 0.9 = 0.018$
- $P(\overline{C} \cap P) = P(\overline{C}) \ P(P) = 0.98 \times 0.05 = 0.049$
- $P(P) = 0.018 + 0.049 = 0.067$

$\Large P(C | P) = \frac{P(P | C) \ P(C)}{P(P)} = \frac{0.018}{0.067} = 0.269$
